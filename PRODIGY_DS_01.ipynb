```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
data_size = 1000

data = {
    'Age': np.random.randint(18, 65, data_size),
    'Gender': np.random.choice(['Male', 'Female', 'Non-binary'], data_size, p=[0.48, 0.48, 0.04])
}
df = pd.DataFrame(data)

print("Sample Dataset Head:")
display(df.head())
```

Now, let's visualize the distribution of the categorical variable 'Gender' using a bar chart.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Distribution of Gender in the Sample Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Next, let's visualize the distribution of the continuous variable 'Age' using a histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Age in the Sample Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
